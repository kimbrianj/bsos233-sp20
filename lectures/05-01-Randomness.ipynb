## Conditional Statements ##

**Scenario:** Suppose you play a betting game in which you win a dollar if a die roll comes up 5 or 6, lose a dollar if the die roll comes up 1 or 2, and tie on a 3 or 4. 

```python
x = 20
if x <= 2:
    print(-1)
elif x <= 4:
    print(0)
if x <= 6:
    print(1)
else:
    print('die has only six sides')
```

```python
def one_bet(x):
    #print('My die roll was',x)
    if x <= 2:
        return -1
    elif x <= 4:
        return 0
    elif x <= 6:
        return 1
    else:
        print('My die only has six faces.')
```

```python
one_bet(5)
```

```python
one_bet(15)
```

### Simulate the Number of Heads in 100 Tosses ###

```python
coin = make_array('heads', 'tails')
```

```python
np.random.choice(coin, 100)
```

```python
sum(np.random.choice(coin, 100) == 'heads')
```

```python
# Simulate one outcome

def num_heads():
    return sum(np.random.choice(coin, 100) == 'heads')
```

```python
# Decide how many times you want to repeat the experiment

repetitions = 10000
```

```python
# Simulate that many outcomes

outcomes = make_array()

for i in np.arange(repetitions):
    outcomes = np.append(outcomes, num_heads())
```

## Fair Coin or not?

Flipped a coin 75 times, got 47 heads. Is this a fair coin?

Initial Assumption: This is a fair coin.

```python
# Simulate one outcome
def num_heads(num_flips):
    coin = make_array('heads','tails')
    return sum(np.random.choice(coin, num_flips) == 'heads')
```

```python
num_heads(75)
```

```python
repetitions = 100000
outcomes = make_array()

for i in np.arange(repetitions):
    outcomes = np.append(outcomes, num_heads(75))
```

```python
heads = Table().with_column('Heads', outcomes)
```

```python
heads.sample(heads.num_rows,with_replacement = False)
```

```python
heads.hist()
```

```python
repetitions * .025
```

```python
repetitions * .975
```

Roll a die 50 times, what is the mean of those 50 rolls?

```python
die_faces = np.arange(1, 7)
die_faces

rolls = np.random.choice(die_faces, 50)
rolls.mean()
```

```python
def roll_50():
    rolls = np.random.choice(die_faces, 50)
    return rolls.mean()
```

```python
repetitions = 10000

outcomes = make_array()

for i in np.arange(repetitions):
    outcomes = np.append(outcomes, roll_50())
```

```python
Table().with_column('Outcome', outcomes).hist()
```
